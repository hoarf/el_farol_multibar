```python
from wlu import Experiment
import numpy as np
```

# Experiment WITH WLU:

## #1) thresholds=[0.2,0.3,0.2,0.1], nr_agents=51, use_wlu=True, alpha=.01, p=.3

```python
na = 51

experiment = Experiment(thresholds=[0.2,0.3,0.2,0.1], nr_agents=na, debug=True, use_wlu=True, alpha=.01, p=.3)

weekly_attendance_data, world_utilities, agent_q_values = experiment.run()
```

### Experiment #1 Results:

#### Agent Q-Values

##### Q-Values for 3 random agents

```python
chosen_agents_ix = np.random.choice(na, 6)
print "Chosen agents: %s" % chosen_agents_ix[0:3]
experiment.plot_q_values(agent_q_values[chosen_agents_ix[0:3]])
```

##### Q-Values for the first 3 agents

```python
experiment.plot_q_values(agent_q_values[:3])
```
