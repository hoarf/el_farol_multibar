# Experiment WITH WLU:

```python
na = 101

experiment = Experiment(nr_weeks=1000, thresholds=[0.3,0.5], nr_agents=na, debug=False,
                        use_wlu=True, alpha=.2, p=.5, decay="partial", init_q_value="rand")

weekly_attendance_data, world_utilities, agent_q_values,rewards, actions = experiment.run()
```

# Experiment WITHOUT WLU

## #2) thresholds=[0.3,0.5], nr_agents=101, use_wlu=False, alpha=.2, p=.5, rand_q_values

```python
na = 101
t = [0.3,0.5]
nw = 5000

experiment = Experiment(nr_weeks = nw, thresholds=t, nr_agents=na,
                        debug=False, use_wlu=False, alpha=.2, p=.5, 
                        decay="partial", init_q_value="rand")

it = 2
nr_actions  = len(t) + 1
weekly_attendance = np.zeros((it, nr_actions,nw))
world_utilities = np.zeros((it, nw))
agent_q_values = np.zeros((it, na, nr_actions, nw))
rewards = np.zeros((it, na, nr_actions, nw))
actions = np.zeros((it, na, nw))

for i in xrange(it):
  weekly_attendance[i], world_utilities[i], agent_q_values[i], rewards[i], actions[i] = experiment.run()  
```

### Experiment #2 Results:

#### Agent Q-Values

##### Q-Values for 2 random agents

```python
chosen_agents_ix = np.random.choice(na, 2)
print "Chosen agents: %s" % chosen_agents_ix
experiment.plot_q_values(np.mean(agent_q_values[:,chosen_agents_ix], axis=0))
```

```python
experiment.plot_rewards(np.mean(rewards[:,chosen_agents_ix], axis=0))
```

```python
experiment.plot_attendances(np.mean(weekly_attendance,axis=0))
```
