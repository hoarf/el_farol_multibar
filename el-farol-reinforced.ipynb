```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
MAXBAR = 2
MAXREWARD = 1000.0
MAXACTIONS = MAXBAR+1
GREEDY = 0
EXPLORE = 1 
ALPHA = .2
MAX_METHOD = 0
RANDOM_METHOD = 1

class Agent:
  
  # Does the agent remember the utility of a bar forever? No: 'No State principle'
  # What to do when every bar have the same utility? Pick one randomly

  def __init__(self):
    self.choices = xrange(MAXACTIONS)
    self.action_q_values = np.random.random_sample(3)*1000
    self.action_choice_count = np.zeros(3)
    
  def __repr__(self):
    return repr(self.actions_utilities)
  
  def chose_action(self, behaviour):
    p = [.5, .5] if behaviour == EXPLORE else [1, 0]
    method = np.random.choice([MAX_METHOD,RANDOM_METHOD], p=p)
    if method == MAX_METHOD:
      best_actions = np.argwhere(self.action_q_values == np.max(self.action_q_values)).flatten()
      self.action = np.random.choice(best_actions)
    else:
      self.action = np.random.choice(self.choices)
    self.action_choice_count[self.action] += 1
  
  def update_utilities(self, rewards):
    r = rewards[self.action]
    q_a = self.action_q_values[self.action]
    n_a = self.action_choice_count[self.action]
    m = np.max(self.action_q_values)
    self.action_q_values[self.action] = q_a + ALPHA*n_a*(r + m - q_a)
    
class World:

  def __init__(self, nr_agents=100):
    self.attendences = np.zeros(MAXACTIONS) 
    self.agents = [ Agent() for x in xrange(nr_agents) ]
    self.thresholds = [.3, .5] # len == MAXBAR
    self.week = 0

  def __repr__(self):
    return repr(self.agents)
    
  def get_reward(self, attendance, threshold):
    a = ((np.float64(attendance)/len(self.agents)-np.float64(threshold))**2)*MAXREWARD #DIFF
    b = a**2
    c = MAXREWARD/np.exp(b)
    return c
    
  def step(self, behaviour=EXPLORE):
    self.attendences = np.zeros(MAXACTIONS) 
    self.bar_results = np.zeros(MAXBAR) 
    self.rewards = np.zeros(MAXACTIONS) 
    
    for agent in self.agents:
      agent.chose_action(behaviour)
      self.attendences[agent.action] += 1
    
    for bar in xrange(MAXBAR):
      self.bar_results[bar] = 1 if self.attendences[bar+1]/len(self.agents) <= self.thresholds[bar] else 0
    
    home_good = np.count_nonzero(self.bar_results) == 0
    
    for action in xrange(MAXACTIONS):
      if action == 0:
        self.rewards[action] = MAXREWARD/MAXBAR if home_good else 0
      else:
        self.rewards[action] = self.bar_results[action-1]*self.get_reward(self.attendences[action], self.thresholds[action-1])
    
    for agent in self.agents:
      agent.update_utilities(self.rewards)

    self.week += 1
      
def ElFarol(wks=5000, nr_agents=100):
  weeks = np.zeros((3,wks,1))
  w = World(nr_agents)
  for week in xrange(wks/2):
    weeks[0][w.week] = w.attendences[0]/len(w.agents)
    weeks[1][w.week] = w.attendences[1]/len(w.agents)    
    weeks[2][w.week] = w.attendences[2]/len(w.agents)
    w.step()

  for week in xrange(wks/2):
    weeks[0][w.week] = w.attendences[0]/len(w.agents)
    weeks[1][w.week] = w.attendences[1]/len(w.agents)    
    weeks[2][w.week] = w.attendences[2]/len(w.agents)
    w.step(GREEDY)
  print "Done week %i" % w.week
  return weeks
```

```python
wks = 500
weeks = ElFarol(wks, nr_agents=10)
x = linspace(0,wks-1,wks)

f, axis = plt.subplots(nrows=3)
axis[0].scatter(x, [weeks[0][k] for k in x], s=1)
axis[0].set_title('Staying home')
axis[0].set_ylim(0,1)
axis[0].set_xlim(0,wks)
axis[0].set_ylabel("Attendance")
axis[0].set_xlabel("Weeks")
axis[0].set_yticks(np.arange(0,1,.1))

axis[1].scatter(x, [weeks[1][k] for k in x], s=1)
axis[1].set_title('bar0')
axis[1].set_ylim(0,1)
axis[1].set_xlim(0,wks)
axis[1].set_ylabel("Attendance")
axis[1].set_xlabel("Weeks")
axis[1].set_yticks(np.arange(0,1,.1))

axis[2].scatter(x, [weeks[2][k] for k in x], s=1)
axis[2].set_title('bar1')
axis[2].set_ylim(0,1)
axis[2].set_xlim(0,wks)
axis[2].set_ylabel("Attendance")
axis[2].set_xlabel("Weeks")
axis[2].set_yticks(np.arange(0,1,.1))

f.tight_layout()
```
