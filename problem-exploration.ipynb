### Analisys of th reward function for t=[0.3,0.5], nr_agents = 101

```python
diff = np.linspace(-1.0,1.0,1000)

def rew(diff):
  b = (((diff-0.5)**2)*1000)**2
  return 1000/np.exp(b) if b < 40 else 0
y = [ rew(x) for x in diff ]
ax = subplot()
ax.plot(diff,y)
ax.set_xlim(0.3,0.7)
for y in xrange(102):
  k = (y/101.0)
  if rew(k) > 1.0 and k <= .5:
    print "Relative attendance: %s Total attendance: %s Reward: %s, " %  (k, y, rew(k))
    ax.axvline(x=k,c='g')
```
