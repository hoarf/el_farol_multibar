### Analisys of th reward function for t=[0.3,0.5], nr_agents = 101

```python
diff = np.linspace(-1.0,1.0,1000)

def rew(diff):
  b = (((diff-0.5)**2)*1000)**2
  return 1000/np.exp(b) if b < 40 else 0
y = [ rew(x) for x in diff ]
ax = subplot()
ax.plot(diff,y)
ax.set_xlim(0.3,0.7)
for y in xrange(102):
  k = (y/101.0)
  if rew(k) > 1.0 and k <= .5:
    print "Relative attendance: %s Total attendance: %s Reward: %s, " %  (k, y, rew(k))
    ax.axvline(x=k,c='g')
```

```python
diff = np.linspace(-1.0,1.0,1000)

def rew(a, t):
  b = (((a-t)**2)*1000)**2
  return 1000/np.exp(b) if b < 40 else 0
y = [ rew(x,0.3) for x in diff ]
ax = subplot()
ax.set_xlim(-0.1,.4)
ax.plot(diff,y)
stay_log = []
for nr_stay in xrange(102):
  for nr_bar1 in xrange(102-nr_stay):
    for nr_bar2 in xrange(102-nr_bar1):
      if nr_stay + nr_bar1 + nr_bar2 == 101:
        a1 = (nr_bar1/101.0)
        a2 = (nr_bar2/101.0)
        if a1 > .3 and a2 > .5:
          stay_log.append(nr_stay)
          ax.axvline(x=nr_stay/101.0,c='g')
print unique(stay_log)
```
