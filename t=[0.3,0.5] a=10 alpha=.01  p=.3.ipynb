```python
from wlu import Experiment
```

# Experiment WITH WLU:

## #1) thresholds=[0.3,0.5], nr_agents=10, use_wlu=True, alpha=.01, p=.3

```python
experiment = Experiment(thresholds=[0.3,0.5], nr_agents=10, debug=False, use_wlu=True, alpha=.01, p=.3)

weekly_attendance_data, world_utilities, agent_q_values, rewards = experiment.run()
```

### Experiment #1 Results:

##### Agents 1-2

```python
experiment.plot_q_values(agent_q_values[0:2])
```

```python
experiment.plot_rewards(rewards[0:2])
```

```python
experiment.plot_attendances(weekly_attendance_data)
```
